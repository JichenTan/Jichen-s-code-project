The following salary data (in USD $) are from [Financial Times 2017](http://rankings.ft.com/businessschoolrankings/global-mba-ranking-2017) and [2016](http://rankings.ft.com/businessschoolrankings/global-mba-ranking-2016).

```
dataset = [ 
 [2016, 1, 'Insead', 166510],
 [2016, 2, 'Harvard', 172501],
 [2016, 3, 'LBS', 154150],
 [2016, 4, 'Wharton', 177877],
 [2016, 5, 'Stanford', 185939],

 [2017, 1, 'Insead', 167657],
 [2017, 2, 'Stanford', 195322],
 [2017, 3, 'Wharton', 181634],
 [2017, 4, 'Harvard', 178113],
 [2017, 5, 'Cambridge', 164462]
];

import pandas as pd
import matplotlib.pyplot as plt
df = pd.DataFrame(dataset, columns=['Year', 'Rank', 'University', 'Salary'])
df
```

```
#fig, ax = plt.subplots(figsize=(16, 9))
def draw_barchart(Year):
    dff = df[df['Year'].eq(Year)].sort_values(by='Salary', ascending=True)
    
    #plot a horizontal bar chart 
    ax.barh(dff['University'],dff['Salary'], color=['lightblue','coral'])
  
    #list the numbers
    for i, (Salary, University) in enumerate(zip(dff['Salary'], dff['University'])):
      ax.text(Salary,i,Salary,ha='right',fontweight='bold',size=20,va='bottom')
      ax.text(Salary, i-.25, University,  ha='right',fontweight='bold',size=20)
        
    #polish the axis

    plt.ylabel("University",size=20)
    plt.xlabel("Salary($)",size=20)
    plt.title("Weighted Salary of MBA Program",size=20)
    ax.text(1, 0.5, Year, transform=ax.transAxes, color='#777777', size=30, ha='right', weight=800)

    #plt.show()

#draw_barchart(2017)
```

```
from matplotlib import animation, rc

fig, ax = plt.subplots(figsize=(16, 9))
animator = animation.FuncAnimation(fig, draw_barchart, frames=range(2016,2018))

# show the animation in Colab
rc('animation', html='jshtml')
animator
```
